```python
import pandas as pd
import numpy as np
#import matplotlib as mp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pkmn = pd.read_csv('Pokemon.csv',sep=',')
pkmn.info()
```

```python
sns.jointplot(x='HP', y='Attack',kind='reg', data=pkmn)
```

```python
plt.figure()
#plt.legend(bbox_to_anchor=(1,1))
sns.boxplot(pkmn.drop(['Legendary','Generation','#','Total'],axis=1))
```

```python
sns.boxplot(y='HP',data=pkmn, palette="winter_d")
```

```python
sns.violinplot(data=pkmn.drop(['Generation','Legendary','#','Total'],axis=1))
```

```python
pkmn = pkmn.drop(['Total','#','Generation','Legendary'],axis=1)
```

```python
sns.stripplot(data=pkmn,palette='Set3')
```

```python
pkmn= pd.melt(pkmn,id_vars=['Name','Type 1','Type 2'],var_name='Stat' )
pkmn.head()
```

```python
plt.figure(figsize=(15,10))
sns.swarmplot(x='Stat',y='value',data=pkmn,hue='Type 1',size=7,split=True)
```

```python
plt.figure(figsize=(12,10))
plt.ylim(0,275)
sns.swarmplot(x='Stat',y='value',data=pkmn,hue='Type 1',split=True,size=7)
plt.legend(bbox_to_anchor=(1,1),loc=2,borderaxespad=0.)
```

```python
pkmn=pd.read_csv('Pokemon.csv')
#pkmn=pd.melt(pkmn.drop(['Legendary','Type 1'],axis=1),id_vars=['Name','Type 2','Generation'],var_name='Stats')
pkmn= pd.melt(pkmn.drop(['Legendary','Type 1','#','Total'],axis=1),id_vars=['Name','Type 2','Generation'],var_name='Stat')
pkmn.head()
```

```python
sns.set_style('whitegrid')
with sns.color_palette('Spectral', desat=.9):
    plt.figure(figsize=(12,10))
    plt.ylim(0,275)
    sns.swarmplot(x='Stat',y='value',data=pkmn, hue='Generation', split=True, size=7)
    plt.legend(bbox_to_anchor=(1,1),loc=2,borderaxespad=0.)
```
